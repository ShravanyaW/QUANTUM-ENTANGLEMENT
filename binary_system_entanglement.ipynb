```python
    import matplotlib.pyplot as plt
    from qiskit import QuantumCircuit
    from qiskit_aer import Aer
    from qiskit import transpile
    from qiskit.visualization import plot_histogram, circuit_drawer
    from qiskit.quantum_info import Statevector

    # Quantum Entanglement Project using Qiskit

    import matplotlib.pyplot as plt

    # Create a 2-qubit quantum circuit
    qc = QuantumCircuit(2)

    # Apply Hadamard gate to qubit 0 to create superposition
    qc.h(0)

    # Apply CNOT gate to entangle qubit 0 and qubit 1
    qc.cx(0, 1)

    # Draw the circuit
    qc.draw(output='mpl')
    plt.title("Quantum Circuit for Entanglement")
    plt.show()

    # Simulate the statevector
    state = Statevector.from_instruction(qc)
    state.draw('bloch')
    plt.suptitle("Bloch Spheres of Entangled Qubits")
    plt.show()

    # Measure the qubits
    qc.measure_all()
    # Set backend as qasm_simulator
    backend = Aer.get_backend('qasm_simulator')

    # Transpile the circuit for the backend
    transpiled_circuit = transpile(qc, backend)

    # Run the transpiled circuit
    job = backend.run(transpiled_circuit, shots=1024)
    result = job.result()
    counts = result.get_counts()


    # Simulate the measurement
    #backend = Aer.get_backend('qasm_simulator')
    #result = execute(qc, backend, shots=1024).result()
    #counts = result.get_counts()
    print(counts)
    bell_counts = {key: value for key, value in counts.items() if key in ['00', '11']}
    plot_histogram(bell_counts)

    plt.title("Measurement Results: Bell State |00> and |11>")
    plt.xlabel("Measured State")
    plt.ylabel("Counts / Probability")
    plt.xticks([0, 1], ['|00', '|11'])
    plt.show()
```
